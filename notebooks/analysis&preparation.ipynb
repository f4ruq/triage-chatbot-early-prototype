```python
df = pd.read_csv("data/Final_Augmented_dataset_Diseases_and_Symptoms.csv")  
symptom_cols = df.columns[1:]  

def convert_symptoms_to_text(row):
    return ", ".join([col for col in symptom_cols if row[col] == 1])

df["text"] = df.apply(convert_symptoms_to_text, axis=1)
```

```python
df["text"]
```

```python
unique_labels = df["diseases"].unique()

label2id = {label: idx for idx, label in enumerate(unique_labels)}
id2label = {idx: label for label, idx in label2id.items()}

df["label"] = df["diseases"].map(label2id)
```

```python
df_ready = df[["text", "label"]]
df_ready.to_csv("bert_ready.csv", index=0)
```
